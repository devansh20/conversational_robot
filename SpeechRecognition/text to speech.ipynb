```python
from gtts import gTTS 
import os
```

```python
text = "Hello Everyone"
```

```python
language = 'en'
```

```python
speech = gTTS(text = text, lang = language, slow = False)
```

```python
speech.save("text.mp3")
```

```python
os.system("start text.mp3")
```

```python
import IPython.display as ipd
ipd.Audio('./text.mp3')
```
